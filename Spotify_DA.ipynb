```python
data_tracks["duration"]= data_tracks["duration_ms"].apply(lambda x: round(x/1000))
```

```python
data_tracks.drop("duration_ms", inplace = True, axis =1)
```

```python
data_tracks.duration.head()
```

```python
#drop the unrequired columns and we will use pearson corr method
correlation_data = data_tracks.drop(["key", "mode", "explicit"],axis = 1).corr(method = "pearson")
plt.figure(figsize=(14,6))
```

```python
heatmap = sns.heatmap(correlation_data, annot =True, fmt =".1g", vmin = -1, vmax = 1, center =0, cmap = "inferno", linewidths =1, linecolor = "Gray")
heatmap.set_title("Correlation HeatMap Between The Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
#sample the data
sample_data = data_tracks.sample(int(0.004*len(data_tracks)))
print(len(sample_data))
```

```python
#recression plot between loudness and energy
plt.figure(figsize=(11,6))
sns.regplot(data = sample_data, y ="loudness", x = "energy",color = 'y').set(title = "Loudness Vs Energy Correlation")
```

```python
#Above plot shows that the energy increases the loudness of the song increases and vice versa
```

```python
#recression plot between popularity Vs acousticness
plt.figure(figsize=(11,6))
sns.regplot(data = sample_data, y ="popularity", x = "acousticness",color = 'r').set(title = "popularity Vs acousticness Correlation")
```
