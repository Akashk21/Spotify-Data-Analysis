```python
popular_most = data_tracks.query('popularity>90',inplace = False ).sort_values('popularity', ascending = False)
popular_most[:10]
```

```python
data_tracks.set_index("release_date", inplace = True)  #inplace means if u want to change the original
```

```python
data_tracks.index = pd.to_datetime(data_tracks.index)
```

```python
data_tracks.head()
```

```python
#index Location Method
data_tracks[["artists"]].iloc[20] #iloc for searching at the index
```

```python
data_tracks.head()
```

```python
data_tracks["duration"]= data_tracks["duration_ms"].apply(lambda x: round(x/1000))
```

```python
data_tracks.drop("duration_ms", inplace = True, axis =1)
```

```python
data_tracks.duration.head()
```

```python
#drop the unrequired columns and we will use pearson corr method
correlation_data = data_tracks.drop(["key", "mode", "explicit"],axis = 1).corr(method = "pearson")
plt.figure(figsize=(14,6))
```

```python
heatmap = sns.heatmap(correlation_data, annot =True, fmt =".1g", vmin = -1, vmax = 1, center =0, cmap = "inferno", linewidths =1, linecolor = "Gray")
heatmap.set_title("Correlation HeatMap Between The Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
#sample the data
sample_data = data_tracks.sample(int(0.004*len(data_tracks)))
print(len(sample_data))
```

```python
#recression plot between loudness and energy
plt.figure(figsize=(11,6))
sns.regplot(data = sample_data, y ="loudness", x = "energy",color = 'y').set(title = "Loudness Vs Energy Correlation")
```

```python
#Above plot shows that the energy increases the loudness of the song increases and vice versa
```

```python
#recression plot between popularity Vs acousticness
plt.figure(figsize=(11,6))
sns.regplot(data = sample_data, y ="popularity", x = "acousticness",color = 'r').set(title = "popularity Vs acousticness Correlation")
```

```python
data_tracks['dates'] = data_tracks.index.get_level_values('release_date')
data_tracks.dates = pd.to_datetime(data_tracks.dates)
years = data_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year ")
```

```python
total_dr=data_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=years, y =total_dr, ax = ax , errwidth = False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

```python
total_dr=data_tracks.duration
sns.set_style(style="whitegrid")
fig_dims = (10,5)
fig, ax = plt.subplots(figsize = fig_dims)
fig= sns.lineplot(x=years, y =total_dr, ax = ax).set(title="Year vs Duration")
plt.xticks(rotation=60)
```
